### Profanity filter

```python
system_prompt = """You are a profanity filter that checks if the human prompt contains any wrong question. There could be "What ifs" but not something that is harmful to human society. Be relaxed, unless you really think that is morally wrong and could have bad consequences.
If the prompt is harmful, return 1. If not, return 0. Don't return anything else.
"""
```

### Prompt and audio generation

```python
system_prompt = """I am creating a video explaining a concept/subject to the general people curious to learn about it. You have to do 2 things:

1. Generate prompts for a video generator to generate videos to generate some easy-to-learn videos explaining concepts.
2. Give some text which AI will use alongside the video(synced) to explain the concept.
3. Each video length will be short. Since you can't explain all concepts. Create multiple video prompts that explain one sub concept.

Some notes: The video should be relevant to the concept you are explaining so choose voiceiver text and prompt accordingly and carefully.
Be super creative with video prompt but use daily life examples. Nothing too complex scene. Also, remember AI is not good in adding some textual information to the video, so don't ask to add some textual information in the video.
explain how things look like, how they are connected and the activity being performed by each. don't say visualize. just tell the scene. explain the scene as much as possible. it should all be in one sentence. no brackets.
Generate as many quantity video prompts as you want to explain the concept fully, but length of number of prompts than equal to 50. Finally, return the response as a json in format.

Stick to this json format and return nothing else. Very strict. My program breaks if you return something else:

{

  prompts: [add prompts],
  voiceovertext: [add voiceover texts]
}
"""
```

```python
completion = client.chat.completions.create(
  model="gpt-4o",
  messages=[
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": input}
  ],
)
```

```python
import torch
from diffusers import CogVideoXPipeline
from diffusers.utils import export_to_video

pipe = CogVideoXPipeline.from_pretrained(
    "THUDM/CogVideoX-5b",
    torch_dtype=torch.bfloat16
)

pipe.enable_model_cpu_offload()
pipe.vae.enable_tiling()
```

```python
for i in range(len(response['prompts'])):
    print(response['prompts'][i])
    prompt = response['prompts'][i]
    video = pipe(
        prompt=prompt,
        num_videos_per_prompt=1,
        num_inference_steps=50,
        num_frames=49,
        use_dynamic_cfg=True,
        guidance_scale=7,
        generator=torch.Generator(device="cuda").manual_seed(42),
    )

    export_to_video(video.frames[0], f"video{i}.mp4")
```

## Stiching everything Together

```python
!virtualenv stitching
!source stitching/bin/activate
```

```python
from moviepy.editor import VideoFileClip, AudioFileClip, concatenate_videoclips, CompositeAudioClip
import moviepy.video.fx.all as vfx
import moviepy.audio.fx.all as afx
from IPython.display import Video

# Lists to keep track of clips for resource management
video_clips = []
audio_clips = []       # To keep track of audio clips to close later
video_file_clips = []  # To keep track of original video clips to close later

for i in range(len(response['prompts'])):
    print(response['prompts'][i])
    # Load the audio and video files
    speech_file = f'speech{i}.mp3'
    video_file = f'video{i}.mp4'

    try:
        audio = AudioFileClip(speech_file)
        audio_clips.append(audio)  # Keep track for closing later
    except Exception as e:
        print(f"Error loading audio file {speech_file}: {e}")
        continue

    try:
        video = VideoFileClip(video_file).without_audio()
        video_file_clips.append(video)  # Keep track for closing later
    except Exception as e:
        print(f"Error loading video file {video_file}: {e}")
        audio.close()  # Close audio if video loading fails
        continue

    # Get durations
    audio_duration = audio.duration
    video_duration = video.duration

    # Repeat video as needed to match audio duration
    num_repeats = int(audio_duration // video_duration)
    remainder = audio_duration % video_duration

    video_clips_repeated = [video] * num_repeats
    if remainder > 0:
        video_clips_repeated.append(video.subclip(0, remainder))

    # Concatenate the repeated video clips
    repeated_video = concatenate_videoclips(video_clips_repeated)
    repeated_video = repeated_video.set_duration(audio_duration)

    # Set audio to the video
    repeated_video = repeated_video.set_audio(audio)

    # Apply fade-in and fade-out to the video (not audio)
    fade_duration = 0.5  # Adjust the fade duration as needed
    if i > 0:
        # Apply fade-in to the beginning of the clip (except the first clip)
        repeated_video = repeated_video.fx(vfx.fadein, fade_duration)
    if i < 9:
        # Apply fade-out to the end of the clip (except the last clip)
        repeated_video = repeated_video.fx(vfx.fadeout, fade_duration)

    # Append the processed clip to the list
    video_clips.append(repeated_video)

    # Do not close the audio and video clips here, as they are still being used

# Concatenate all video clips sequentially
final_clip = concatenate_videoclips(video_clips, method='compose')


background_music_file = 'background_music.mp3'  # Replace with your background music file
try:
    background_music = AudioFileClip(background_music_file)
except Exception as e:
    print(f"Error loading background music file '{background_music_file}': {e}")
    background_music = None

# Add background music if available
if background_music:
    # Get the duration of the final video
    final_duration = final_clip.duration

    # Adjust background music to match the final video's duration
    if background_music.duration < final_duration:
        # Loop the background music
        background_music = afx.audio_loop(background_music, duration=final_duration)
    else:
        # Trim the background music
        background_music = background_music.subclip(0, final_duration)

    # Reduce background music volume
    background_music = background_music.volumex(0.2)  # Adjust volume as needed (0.2 means 20% of original volume)

    # Combine the background music with the final clip's audio
    final_audio = CompositeAudioClip([final_clip.audio, background_music])

    # Set the combined audio to the final clip
    final_clip = final_clip.set_audio(final_audio)

# Write the output video file
final_clip.write_videofile('output_video_with_music.mp4', codec='libx264', audio_codec='aac')

# Close all clips to free resources
for clip in audio_clips:
    clip.close()
for clip in video_file_clips:
    clip.close()
for clip in video_clips:
    clip.close()
final_clip.close()
```
